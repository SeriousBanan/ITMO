```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [16, 12]
```

```python
DOTS = [{
    "val": 100,
    "x": 40,
    "y": 40
},{
    "val": 100,
    "x": 60,
    "y": 60
}]

SIZE = (100, 100)
STEPS = 10000
```

```python
field = np.zeros(SIZE)

for dot in DOTS:
    field[dot["x"], dot["y"]] = dot["val"]
X = np.arange(0, SIZE[0], 1)
X = np.arange(0, SIZE[0], 1)
v_sum = np.zeros(STEPS)
```

```python
for i in range(1, STEPS):
    v_sum[i] = field.sum()
    print(f"\rV_div: {v_sum[i] - v_sum[i - 1]:.2} [{i:>4}/3000]", end="")
    if v_sum[i] - v_sum[i - 1] < 5:
        break
    for x in X:
        for y in Y:
            if x in (0, SIZE[0] - 1) or y in (0, SIZE[1] - 1) or (x, y) in [(dot["x"], dot["y"]) for dot in DOTS]:
                continue
            field[x][y] = (field[x - 1][y] + field[x + 1][y] + field[x][y - 1] + field[x][y + 1]) / 4
```

```python
fig = plt.figure()

ax = plt.axes(projection='3d')
ax.plot_surface(np.meshgrid(X, Y)[0] , np.meshgrid(X, Y)[1], field, rstride=1, cstride=1,
                cmap='viridis')
ax.set_title('surface')
```
