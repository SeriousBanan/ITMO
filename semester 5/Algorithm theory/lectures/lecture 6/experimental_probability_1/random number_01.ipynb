### Два примера на тему 
### линейные random number generators (RNG)

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
#   matplotlib.style.use('ggplot')
%matplotlib inline
```

####   Линейный RNG.  Пример 1  <br>

$x_n = 4809922 \cdot x_{n−1} \; mod \; 60466169$  <br>
$u_n = x_n / 60466169$

```python
#  Число экспериментов
n_exp = 1000


#  Готовлю векторы для последовательностей зерен и (псевдо)случайных чисел
x = [-9999]*n_exp
u = [-9999]*n_exp

#  Зерно датчика и первое случайное число
x[0] = 12345678
u[0] = 12345678/60466169

```

```python
#  Последовательность случайных чисел длины n_exp

for i in range(1, n_exp):
    #    
    x[i] = (4809922 * x[i-1]) % 60466169
    u[i] = x[i] / 60466169 
```

```python
#  мало ли что...
x_1 = u.copy()
y_1 = u.copy()
```

```python
#  Создаю пары соседних случайных чисел
del x_1[-1]
del y_1[0]

```

```python
#  Координаты точек - пары соседних случайных чисел
plt.scatter(x_1, y_1)
```

####   Линейный RNG.  Пример 2  <br>

$x_n = 12 \cdot x_{n−1} \; mod \; 101$  <br>
$u_n = x_n / 101$

```python
#  Число экспериментов
n_exp = 1000

#  Готовлю вектор для последовательности зерен и (псевдо)случайных чисел
x = [-9999]*n_exp
u = [-9999]*n_exp

#  Зерно датчика и первое случайное число
x[0] = 12345678
u[0] = 12345678/60466169

```

```python
#  Последовательность случайных чисел длины n_exp

for i in range(1, n_exp):
    #    
    x[i] = (12 * x[i-1]) % 101
    u[i] = x[i] / 101

#  мало ли что...
x_1 = u.copy()
y_1 = u.copy()

#  Создаю пары соседних случайных чисел
del x_1[-1]
del y_1[0]

```

```python
plt.scatter(x_1, y_1)
```
