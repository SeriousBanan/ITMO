```python
df = pd.read_csv('Chicago_hotels.csv', sep=';', decimal=',', na_values = " ")
df.head()
```

```python
df = df[0:112]
```

```python
df.tail(15)
```

```python
y = df["x2"]
```

```python
y.plot()
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
fit1 = ExponentialSmoothing(np.asarray(y) ,seasonal_periods=12 ,trend='additive', seasonal='multiplicative', damped = True).fit()
```

```python
len(y)
```

```python
fitted_ = fit1.predict(0, 121)
```

```python
fitted_
```

```python
plt.figure(figsize=(16,8))
plt.plot(fitted_, label='Holt_Winters')
plt.plot(np.asarray(y), label='y')
plt.legend(loc='best')
plt.show()
```
